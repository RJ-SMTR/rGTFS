```python
import basedosdados as bd
import geopandas as gpd
from shapely.geometry import Point
import folium

```

```python
gtfs_tables = ['shapes', 'trips', 'stop_sequences', 'stops']
gtfs = dict()
for tb in gtfs_tables:
    gtfs[tb] = bd.read_table(
        dataset_id='br_rj_riodejaneiro_gtfs',
        table_id=tb,
        query_project_id='rj-smtr',
        billing_project_id='rj-smtr'
    )
```

```python
stops = bd.read_sql(
    """
    SELECT 
        t1.*,
        t2.stop_lon longitude,
        t2.stop_lat latitude,
        ST_GEOGPOINT(t2.stop_lon, t2.stop_lat) point
    FROM rj-smtr.br_rj_riodejaneiro_gtfs.stop_sequences t1
    JOIN rj-smtr.br_rj_riodejaneiro_gtfs.stops t2
    ON t1.stop_id = t2.stop_id
    """,
    billing_project_id='rj-smtr'
)
```

```python
selected = stops.query('route_id == "46"').query('direction_id == 0')
```
